```python
# prompt: give me a list of datatypes with th example and short explain

# Numeric Types

# int: Integer numbers (whole numbers)
age = 30  # Example
print(f"Age: {age}, Type: {type(age)}") # Output: Age: 30, Type: <class 'int'>

# float: Floating-point numbers (numbers with decimal points)
price = 99.99 # Example
print(f"Price: {price}, Type: {type(price)}") # Output: Price: 99.99, Type: <class 'float'>

# complex: Complex numbers (numbers with a real and imaginary part)
z = 2 + 3j # Example
print(f"Complex Number: {z}, Type: {type(z)}") # Output: Complex Number: (2+3j), Type: <class 'complex'>


# Text Type

# str: String (sequence of characters)
name = "Alice"  # Example
print(f"Name: {name}, Type: {type(name)}") # Output: Name: Alice, Type: <class 'str'>


# Sequence Types

# list: Ordered, mutable (changeable) sequence of items
my_list = [1, 2, 3, "apple", 4.5] # Example: Mixed data types
print(f"List: {my_list}, Type: {type(my_list)}") # Output: List: [1, 2, 3, 'apple', 4.5], Type: <class 'list'>

# tuple: Ordered, immutable (unchangeable) sequence of items
my_tuple = (1, 2, 3, "banana") # Example
print(f"Tuple: {my_tuple}, Type: {type(my_tuple)}") # Output: Tuple: (1, 2, 3, 'banana'), Type: <class 'tuple'>

# range: Represents a sequence of numbers
numbers = range(1, 6) # Example (numbers 1 through 5)
print(f"Range: {list(numbers)}, Type: {type(numbers)}") # Output: Range: [1, 2, 3, 4, 5], Type: <class 'range'>


# Mapping Type

# dict: Dictionary (key-value pairs)
person = {"name": "Bob", "age": 25, "city": "New York"} # Example
print(f"Dictionary: {person}, Type: {type(person)}") # Output: Dictionary: {'name': 'Bob', 'age': 25, 'city': 'New York'}, Type: <class 'dict'>


# Set Types

# set: Unordered collection of unique items
my_set = {1, 2, 3, 3, 4} # Example (duplicates are automatically removed)
print(f"Set: {my_set}, Type: {type(my_set)}") # Output: Set: {1, 2, 3, 4}, Type: <class 'set'>


# frozenset: Immutable version of a set
my_frozenset = frozenset([1, 2, 3, 3, 4]) # Example
print(f"Frozen Set: {my_frozenset}, Type: {type(my_frozenset)}") # Output: Frozen Set: frozenset({1, 2, 3, 4}), Type: <class 'frozenset'>


# Boolean Type

# bool: Boolean values (True or False)
is_active = True # Example
print(f"Is Active: {is_active}, Type: {type(is_active)}") # Output: Is Active: True, Type: <class 'bool'>


# Binary Types

# bytes: Sequence of bytes
byte_data = b"Hello" # Example
print(f"Bytes: {byte_data}, Type: {type(byte_data)}") # Output: Bytes: b'Hello', Type: <class 'bytes'>


# None Type

# NoneType: Represents the absence of a value
empty_variable = None # Example
print(f"Empty variable: {empty_variable}, Type: {type(empty_variable)}")  # Output: Empty variable: None, Type: <class 'NoneType'>
```

```python
# prompt: What are the different core operators in python. please give examples

# Arithmetic Operators
x = 10
y = 5

print(f"Addition: {x + y}")      # Output: 15
print(f"Subtraction: {x - y}")   # Output: 5
print(f"Multiplication: {x * y}") # Output: 50
print(f"Division: {x / y}")      # Output: 2.0
print(f"Floor Division: {x // y}") # Output: 2 (integer division)
print(f"Modulo: {x % y}")        # Output: 0 (remainder)
print(f"Exponentiation: {x ** y}") # Output: 100000 (x to the power of y)

# Comparison Operators
print(f"Equal to: {x == y}")    # Output: False
print(f"Not equal to: {x != y}")  # Output: True
print(f"Greater than: {x > y}")   # Output: True
print(f"Less than: {x < y}")    # Output: False
print(f"Greater than or equal to: {x >= y}") # Output: True
print(f"Less than or equal to: {x <= y}")  # Output: False

# Logical Operators
a = True
b = False

print(f"Logical AND: {a and b}") # Output: False
print(f"Logical OR: {a or b}")  # Output: True
print(f"Logical NOT: {not a}") # Output: False

# Bitwise Operators (operate on binary representations of numbers)
num1 = 10  # Binary: 1010
num2 = 4   # Binary: 0100

print(f"Bitwise AND: {num1 & num2}")  # Output: 0 (Binary: 0000)
print(f"Bitwise OR: {num1 | num2}")   # Output: 14 (Binary: 1110)
print(f"Bitwise XOR: {num1 ^ num2}")  # Output: 14 (Binary: 1110)
print(f"Bitwise NOT: {~num1}")      # Output: -11 (Inverts bits)
print(f"Left shift: {num1 << 1}")    # Output: 20 (Shifts bits to the left)
print(f"Right shift: {num1 >> 1}")   # Output: 5 (Shifts bits to the right)


# Assignment Operators
z = 5
z += 3  # Equivalent to z = z + 3 (z becomes 8)
z -= 2  # Equivalent to z = z - 2 (z becomes 6)
z *= 4  # Equivalent to z = z * 4 (z becomes 24)
# Similar for /=, //=, %=, **=, &=, |=, ^=, <<=, >>=

# Identity Operators (check if variables refer to the same object in memory)
list1 = [1, 2, 3]
list2 = list1
list3 = [1, 2, 3]

print(f"Is list1 the same object as list2?: {list1 is list2}") # Output: True
print(f"Is list1 the same object as list3?: {list1 is list3}") # Output: False (even though they have the same content)

# Membership Operators (check if a value is a member of a sequence)
my_string = "Hello"
print(f"Is 'e' in my_string?: {'e' in my_string}") # Output: True
print(f"Is 'x' not in my_string?: {'x' not in my_string}") # Output: True
```

```python
# prompt: Help us with the core conditional in python using example

# Conditional Statements (if, elif, else)

# Example 1: Checking if a number is positive, negative, or zero
number = 10

if number > 0:
    print("The number is positive")
elif number < 0:
    print("The number is negative")
else:
    print("The number is zero")


# Example 2: Checking if a person is eligible to vote
age = 18

if age >= 18:
    print("You are eligible to vote.")
else:
    print("You are not eligible to vote yet.")


# Example 3:  Nested Conditional Statements
grade = 85  # Example

if grade >= 90:
    print("A")
elif grade >= 80:
    if grade >= 85:  # Nested if
        print("B+")
    else:
        print("B")

elif grade >= 70:
    print("C")
else:
    print("Below C")

#Example 4: Using 'and', 'or', and 'not' with conditional statements
x = 15
y = 20
if x > 10 and y < 25:
    print("Both conditions are true.")
else:
    print("At least one condition is false.")
```

```python
# prompt: give some basic examples of doing loop in python

# For loop (iterating through a sequence)
my_list = [1, 2, 3, 4, 5]

for item in my_list:
  print(item) # Output: 1 2 3 4 5


# For loop with range()
for i in range(5): # Iterates from 0 to 4
  print(i) # Output: 0 1 2 3 4

for i in range(2, 6):  # Iterates from 2 to 5
  print(i)  # Output: 2 3 4 5

for i in range(1, 10, 2): # Iterates from 1 to 9 with a step of 2
  print(i)  # Output: 1 3 5 7 9

# While loop (repeats as long as a condition is True)
count = 0
while count < 5:
  print(count) # Output: 0 1 2 3 4
  count += 1


# Nested loops
for i in range(3):
  for j in range(2):
    print(f"i={i}, j={j}")
```

```python
#create a number guessing game
import random

def play_number_guessing_game():
  #user get onlu 7 chance to guess
  #the numbers are between 1 to 99
  secret_number = random.randint(0, 100)
  Guesses_left = 7

  print("Welcome to the Number Guessing Game!")
  print("I've selected a random number between 0 and 100. Can you guess it?")

  while Guesses_left > 0:
      guess = int(input("Enter your guess: "))

      if guess == secret_number:
          print(f"Congratulations! You guessed the number {secret_number} correctly!")
          break
      elif guess < secret_number:
          print("Too low! Try again.")
          Guesses_left -= 1
          print(f"You have {Guesses_left} guesses left.")
      else:
          print("Too high! Try again.")
          Guesses_left -= 1
          print(f"You have {Guesses_left} guesses left.")


      if Guesses_left == 0:
          print(f"Sorry, you've run out of guesses. The secret number was {secret_number}.")

play_number_guessing_game()
```
